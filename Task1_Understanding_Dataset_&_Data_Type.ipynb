```python
#Titanic Dataset
import pandas as pd
import numpy as np
```

```python
# Load Titanic dataset
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)

# Display first 5 rows
df.head()
```

```python
df.tail()
```

```python
#Dataset Information (Data Types & Null Values)
print("Dataset Shape:", df.shape)
```

```python

df.info()
```

```python
# Statistical Summary (Numerical Data)
df.describe()
```

```python
#Identify Column Types
numerical_features = df.select_dtypes(include=['int64', 'float64']).columns
categorical_features = df.select_dtypes(include=['object']).columns

print("Numerical Features:\n", numerical_features)
print("\nCategorical Features:\n", categorical_features)
```

```python
#Check Unique Values in Categorical Columns
for col in categorical_features:
    print(f"\nColumn: {col}")
    print(df[col].value_counts())
```

```python
df.isnull().sum()
```

```python
target = "Survived"
features = df.drop(columns=[target]).columns

print("Target Variable:", target)
print("Input Features:\n", features)
```

```python
df[target].value_counts()
```

```python
if df.shape[0] > 500:
    print("Dataset is sufficient for basic ML models ")
else:
    print("Dataset is small, may cause overfitting ")
```

```python
#1. Dataset contains both numerical and categorical features.
#2. No major missing values are present.
#3. Target variable is automatically identified using sklearn.
#4. Dataset size is sufficient for basic machine learning models.
#5. Some categorical features have multiple unique values, requiring encoding before modeling.
```

```python
#Students Performance Dataset
import pandas as pd
from sklearn.datasets import fetch_openml
```

```python
# Load Student Performance (Math) dataset from OpenML
data = fetch_openml(data_id=1590, as_frame=True)

df = data.frame
df.head()
```

```python
df.tail()
```

```python
df.shape
```

```python
df.describe()
```

```python
numerical_features = df.select_dtypes(include=['int64', 'float64']).columns
categorical_features = df.select_dtypes(include=['object', 'category']).columns

print("Numerical Features:")
print(numerical_features)

print("\nCategorical Features:")
print(categorical_features)
```

```python
for col in categorical_features:
    print(f"\nColumn: {col}")
    print(df[col].value_counts())
```

```python
df.isnull().sum()
```

```python
target = data.target.name
print("Target column:", target)

# Features
features = df.drop(columns=[target]).columns
print("Number of features:", len(features))
print("Target Variable:", target)
print("Number of Input Features:", len(features))
```

```python
df[target].value_counts()
```

```python
if df.shape[0] >= 300:
    print("Dataset is suitable for Machine Learning ")
else:
    print("Dataset is small and may cause overfitting ")
```

```python
#1. Dataset contains both numerical and categorical features.
#2. No major missing values are present.
#3. Target variable is automatically identified using sklearn.
#4. Dataset size is sufficient for basic machine learning models.
#5. Some categorical features have multiple unique values, requiring encoding before modeling.
```
